## Case 1

$$g(x)=x(\pi-x)$$
$$
v_n(0)=b_n=\frac{1}{\pi} \int_{-\pi}^\pi g(x) \sin (n x) d x
$$
$$
v_n(0)=b_n=\frac{1}{\pi} \int_{-\pi}^\pi x(\pi-x) \sin (n x) d x
$$
Here we use integration by parts to eliminate $x$ from the integrand:
$$
=\frac{2}{\pi}\left[-x(\pi-x) \frac{\cos (n x)}{n}\right]_0^\pi+\frac{2}{\pi} \int_0^\pi(\pi-2 x) \frac{\cos (n x)}{n} d x
$$
Observe that:
$$
\frac{2}{\pi}\left[-x(\pi-x) \frac{\cos (n x)}{n}\right]_0^\pi=0
$$
$$
\frac{2}{\pi}\left[-x(\pi-x) \frac{\cos (n x)}{n}\right]_0^\pi+\frac{2}{\pi} \int_0^\pi(\pi-2 x) \frac{\cos (n x)}{n} d x=\frac{2}{\pi} \int_0^\pi(\pi-2 x) \frac{\cos (n x)}{n} d x
$$
Here again we can use integration by parts to eliminate  $(\pi-2 x)$ from the integrand:
$$
=\frac{2}{\pi}\left[(\pi-2 x) \frac{\sin (n x)}{n^2}\right]_0^\pi+\frac{2}{\pi} \int_0^\pi 2 \frac{\sin (n x)}{n^2} d x
$$
Observe that:
$$
\frac{2}{\pi}\left[(\pi-2 x) \frac{\sin (n x)}{n^2}\right]_0^\pi=0
$$
$$
\frac{2}{\pi}\left[(\pi-2 x) \frac{\sin (n x)}{n^2}\right]_0^\pi+\frac{2}{\pi} \int_0^\pi 2 \frac{\sin (n x)}{n^2} d x=\frac{2}{\pi} \int_0^\pi 2 \frac{\sin (n x)}{n^2} d x=
$$
$$
=\frac{4}{\pi} \int_0^\pi \frac{\sin (n x)}{n^2} d x=\frac{4}{\pi}\left[\frac{-\cos (n x)}{n^3}\right]_0^\pi=\frac{4}{\pi}\left(\frac{1-(-1)^n}{n^3}\right)
$$
\textbf{NOTE:} Symmetry consideration: $g(x)=x(\pi-x)$ so $g(-x)=-g(x)$ $\forall x \in [0, \pi]$.\\\\
Since:
$$
g\left(\frac{x_1}{x_2}\right)=\sum_{n=1}^{\infty} b_n \sin \left(n x_1\right)
$$ 
Where $b_n$ is given by:
$$
b_n=\frac{1}{\pi} \int_{-\pi}^\pi g\left(\frac{x_1}{x_2}\right) \sin \left(n x_1\right) d x_1
$$
Substituting the Fourier series expansion of $g\left(\frac{x_1}{x_2}\right)$ back into the expression for $u(x, \tau)$, we obtain the general solution as:
$$
u(x, \tau)=\frac{4}{\pi} \sum_{n=1}^{\infty}\left(\frac{1-(-1)^n}{n^3}\right) \mathrm{e}^{-n^2 \tau} \sin (n x) .
$$

## Case 2

$$
g(x)=\left\{\begin{array}{l}
x, x \in[0, \pi / 2] \\
0, x \in[\pi / 2, \pi]
\end{array}\right.
$$
$$
v_n(0)=b_n=\frac{1}{\pi} \int_{-\pi}^\pi g(x) \sin (n x) d x
$$
We have an odd function $g(-x)=-g(x)$ as well as $sin(-x)=-sin(x)$. So w can write:
$$=2\frac{1}{\pi} \int_0^\pi g(x) \sin (n x) d x$$
Now since we have a piece wise defined function we have to to split the integral into two parts.
$$
2\frac{1}{\pi} \int_0^{\pi / 2} x \sin (n x) d x+2\frac{1}{\pi} \int_{\pi / 2}^\pi 0 \cdot \sin (n x) d x
$$
This simplifies the integral to solving:
$$
2\frac{1}{\pi} \int_0^{\pi / 2} x \sin (n x) d x
$$
Here again we use integration by parts to eliminate $x$ from the integrand:
$$
\frac{2}{\pi} \int_0^{\pi / 2} x \sin (n x) d x=\frac{2}{\pi}\left[\frac{-x \cos (n x)}{n}\right]_0^{\pi / 2}+\frac{2}{\pi} \int_0^{\pi / 2} \frac{\cos (n x)}{n} d x=
$$
$$
=\frac{2}{\pi}\frac{-\pi \cos \left(\frac{\pi n}{2}\right)}{2 n}+\frac{2}{\pi} \int_0^{\pi / 2} \frac{\cos (n x)}{n} d x=
$$
$$
=-\frac{1}{n} \cos \left(\frac{\pi n}{2}\right)+\frac{2}{\pi}\left[\frac{\sin (n x)}{n^2}\right]_0^{\pi / 2}=
$$
$$
=-\frac{1}{n} \cos \left(\frac{\pi n}{2}\right)+\frac{2}{\pi n^2} \sin \left(\frac{n \pi}{2}\right)
$$
The General solution is again given by the:
$$
u(x, \tau)=\sum_{n=1}^{\infty} b_n e^{-n^2 \tau} \sin (n x)
$$
$$
u(x, \tau)=\sum_{n=1}^{\infty} (-\frac{1}{n} \cos \left(\frac{\pi n}{2}\right)+\frac{2}{\pi n^2} \sin \left(\frac{n \pi}{2}\right)) e^{-n^2 \tau} \sin (n x)
$$
Observe that for different $n$ we have $4$ cases that simplify the expression:\\

$$
\text{if } n \bmod 4 = 0 \Rightarrow -\frac{1}{n}
$$
$$
\text{if } n \bmod 4 = 1 \Rightarrow \frac{2}{\pi n^2}
$$
$$
\text{if } n \bmod 4 = 2 \Rightarrow \frac{1}{n}
$$
$$
\text{if } n \bmod 4 = 3 \Rightarrow -\frac{2}{\pi n^2}
$$

## Comparing Numerical Methods: Fourier Series vs. Finite Difference Method

Forward Difference for Time Derivative:
$$
\frac{u\left(x_n, \tau_{m+1}\right)-u\left(x_n, \tau_m\right)}{\Delta \tau}
$$
Central Difference for Spatial Second Derivative:
$$
\frac{u\left(x_{n+1}, \tau_m\right)-2 u\left(x_n, \tau_m\right)+u\left(x_{n-1}, \tau_m\right)}{\Delta x^2}
$$
Thus we can approximate $\frac{\partial}{\partial \tau} u(x, \tau)-\frac{\partial^2}{\partial x^2} u(x, \tau)=0$ as:
$$
\frac{u\left(x_n, \tau_{m+1}\right)-u\left(x_n, \tau_{m n}\right)}{\Delta \tau}-\frac{u\left(x_{n+1}, \tau_m\right)+u\left(x_{n-1}, \tau_m\right)-2 u\left(x_n, \tau_m\right)}{\Delta x^2}=0
$$
$$
\frac{u\left(x_n, \tau_{m+1}\right)-u\left(x_n, \tau_m\right)}{\Delta \tau}=\frac{u\left(x_{n+1}, \tau_m\right)-2 u\left(x_n, \tau_m\right)+u\left(x_{n-1}, \tau_m\right)}{\Delta x^2}
$$
$$
u\left(x_n, \tau_{m+1}\right)-u\left(x_n, \tau_m\right)=\Delta \tau \frac{u\left(x_{n+1}, \tau_m\right)-2 u\left(x_n, \tau_m\right)+u\left(x_{n-1}, \tau_m\right)}{\Delta x^2}
$$
$$
u\left(x_n, \tau_{m+1}\right)=u\left(x_n, \tau_m\right)+\Delta \tau \frac{u\left(x_{n+1}, \tau_m\right)+u\left(x_{n-1}, \tau_m\right)-2 u\left(x_n, \tau_m\right)}{\Delta x^2}
$$
We remember our conditions:
$$
\begin{array}{llrl}
u(x, 0) & =g(x), & & x \in[0, \pi] \\
u(0, \tau) & =0, & & \tau>0 \\
u(\pi, \tau) & =0, & & \tau>0 .
\end{array}
$$
