### Hedged Portfolio

$$
\Pi=V-\Delta S
$$

#### Components of the Equation:
- **$\Pi$**: Represents the value of the hedged portfolio.
- **$V$**: Represents the value of the option or derivative being hedged.
- **$\Delta$**: Represents the "delta" of the option, which is a measure of the rate of change of the option value with respect to the changes in the price of the underlying asset.
- **$S$**: Represents the price of the underlying asset (e.g., stock, commodity).

#### Understanding the Hedged Portfolio:
This equation shows how a portfolio can be "hedged" to reduce risk. In practice, $V$ is the part of the portfolio that is sensitive to movements in the market (such as an option), and $S$ is the underlying asset which influences the value of $V$. The term $\Delta S$ represents a position in the underlying asset that is adjusted for the sensitivity (delta) of the option.

The idea is to construct the portfolio such that the change in $V$ due to a small change in $S$ is offset by the change in $\Delta S$. This is done by taking a position in the underlying asset that is opposite in direction to the expected movement of the option value:
- If $\Delta$ is positive, the investor would hold a position in the underlying asset equivalent to $\Delta$.
- If $\Delta$ is negative, the investor would short-sell $\Delta$ units of the underlying asset.

#### Practical Implication:
The practical goal of this formula is to create a portfolio where the net exposure to market movements is minimized. By adjusting the composition of $\Delta S$, the portfolio's total value $\Pi$ becomes less sensitive to price fluctuations of the underlying asset $S$. This technique is often used to "neutralize" the risk of price movements in the short term, ensuring that the portfolio's value remains relatively stable regardless of market volatility.

In essence, the hedged portfolio equation provides a mathematical way to balance an investment against its risks.

### Return on Portfolio

$$
d \Pi=d V-\frac{\partial V}{\partial S} d S
$$

#### Components of the Equation:
- **$d \Pi$**: Represents the incremental change in the value of the hedged portfolio.
- **$d V$**: Represents the incremental change in the value of the option or derivative.
- **$\frac{\partial V}{\partial S}$**: Denotes the partial derivative of the option value with respect to the stock price, commonly known as the 'delta' ($\Delta$) of the option.
- **$d S$**: Represents the incremental change in the stock price (the underlying asset).

#### Understanding the Return on the Portfolio:
This equation essentially calculates the net change in the portfolio's value by considering how the value of the derivative $V$ changes and how those changes are offset by changes in the value of the underlying asset $S$, scaled by the delta of the option.

#### Breakdown:
- **$d V$** captures any direct change in the derivative's value due to factors like underlying asset price changes, volatility, time decay (theta), and other Greeks.
- **$\frac{\partial V}{\partial S} d S$** adjusts this change by accounting for the rate at which $V$ is expected to change as $S$ changes. This term basically multiplies the delta of the option ($\Delta$) by the change in the stock price, which represents how much the position in the underlying asset (used to hedge the option) has gained or lost.

#### Practical Implication:
The purpose of this formula is to determine the total differential return of a portfolio that includes both an option and a position in the underlying asset (used to hedge the option position). The term $d V - \frac{\partial V}{\partial S} d S$ reflects the idea that the changes in the option's value are ideally offset by the changes in the hedged position, aiming to stabilize the overall portfolio value against small movements in the stock price.

This formula is foundational in the practice of "delta hedging," where the goal is to make the portfolio delta-neutral, meaning that the portfolio's value is not affected by small price movements of the underlying asset. In a perfectly delta-hedged portfolio, $d \Pi$ would theoretically be zero if only small changes in $S$ occur, meaning there is no net gain or loss due to these changes—thereby isolating the portfolio from risks associated with fluctuations in the underlying asset's price.

### Itô's lemma


$$
d V=\left(\frac{\partial V}{\partial S} \mu S+\frac{\partial V}{\partial t}+\frac{1}{2} \sigma^2 S^2 \frac{\partial^2 V}{\partial S^2}\right) d t+\frac{\partial V}{\partial S} \sigma S d z
$$

#### Components of the Itô's Lemma Equation:
- **$dV$**: Represents the infinitesimal change in the value of a function $V(S, t)$, which in finance, is typically the value of an option.
- **$\frac{\partial V}{\partial S}$**: The partial derivative of $V$ with respect to the stock price $S$ (delta, $\Delta$).
- **$\frac{\partial V}{\partial t}$**: The partial derivative of $V$ with respect to time $t$ (theta, $\Theta$).
- **$\frac{1}{2} \sigma^2 S^2 \frac{\partial^2 V}{\partial S^2}$**: This term accounts for the curvature (or convexity) of the value function with respect to the stock price, incorporating the volatility $\sigma$ of the stock. This second derivative is often referred to as gamma ($\Gamma$).
- **$\mu S$**: The drift component of the stock's return, with $\mu$ representing the expected return of the stock.
- **$\sigma S dz$**: The random shock to the stock price, where $\sigma$ is the stock's volatility and $dz$ represents a standard Brownian motion, indicating the random, unpredictable changes in stock price.

### Explanation of the Equation:
Itô's lemma shows how the option value $V$ changes due to small changes in the stock price $S$ and time $t$. The first part of the equation,
$$
\left(\frac{\partial V}{\partial S} \mu S + \frac{\partial V}{\partial t} + \frac{1}{2} \sigma^2 S^2 \frac{\partial^2 V}{\partial S^2}\right) dt,
$$
captures the expected change in $V$ over an infinitesimal time $dt$, considering both the directional movement due to the drift and the effect of the option's gamma on the volatility of the stock.

The second part,
$$
\frac{\partial V}{\partial S} \sigma S dz,
$$
represents the random fluctuation in $V$ due to the stochastic change in the stock price, scaled by the option's delta. This term integrates the uncertainty or risk (modeled as a Brownian motion) into the option's pricing model.

### Brownian Motion with Drift

$$
d S = \mu S dt + \sigma S dz
$$ 

#### Components of the Equation:
- **$dS$**: Represents the infinitesimal change in the stock price $S$.
- **$\mu$**: The drift coefficient, representing the expected return or average rate of increase of the stock price per unit of time. This can be thought of as the "trend" component of the stock price movement.
- **$S$**: The current stock price.
- **$dt$**: An infinitesimal increment of time.
- **$\sigma$**: The volatility coefficient, which measures the amount of uncertainty or risk associated with the stock price's changes. This coefficient determines the standard deviation of the stock's returns.
- **$dz$**: Represents an increment of a Wiener process (or standard Brownian motion), which models the random fluctuation in the stock price.

#### Explanation of the Equation:
The equation $d S = \mu S dt + \sigma S dz$ is used to model the dynamics of stock prices under the assumption that they follow a stochastic differential equation driven by both deterministic and random influences:
- **$\mu S dt$**: This term accounts for the expected, deterministic growth of the stock price over time, scaled by the current stock price $S$. It reflects the average or expected growth rate of the stock assuming no random fluctuations.
- **$\sigma S dz$**: This term introduces randomness into the model. It represents the unexpected changes in the stock price, influenced by external factors and market sentiment. The term $\sigma S$ scales the volatility to the current price level, and $dz$ brings in the element of randomness, modeling these changes as a normal distribution with mean zero and variance $dt$.

#### Practical Significance:
- **Modeling Stock Prices**: This model is central in financial mathematics, particularly in the pricing of stock options and other derivatives. It underpins the Black-Scholes formula and is used extensively in quantitative finance to model asset paths.
- **Risk Assessment**: By adjusting $\mu$ and $\sigma$, financial analysts can simulate different scenarios and assess the potential risk and return profiles of investments.
- **Market Simulations**: It is also used in Monte Carlo simulations to forecast future stock price movements and to calculate the risk associated with different trading strategies.

### The Simplified Equation

$$
d \Pi=\left(-\frac{\partial V}{\partial t}-\frac{1}{2} \sigma^2 S^2 \frac{\partial^2 V}{\partial S^2}\right) d t
$$

#### Components of the Equation:
- **$d\Pi$**: Represents the infinitesimal change in the value of the hedged portfolio.
- **$\frac{\partial V}{\partial t}$**: The partial derivative of the option value $V$ with respect to time $t$. This term often relates to the "theta" ($\Theta$) of the option, which measures the sensitivity of the option's price to the passage of time.
- **$\sigma^2$**: The volatility squared, reflecting the variance in the underlying asset's returns.
- **$S^2$**: The square of the stock price.
- **$\frac{\partial^2 V}{\partial S^2}$**: The second partial derivative of the option value with respect to the stock price $S$, known as "gamma" ($\Gamma$). Gamma measures the rate of change in delta ($\Delta$) and provides insight into the convexity of the option's value relative to changes in the underlying asset.
- **$dt$**: An infinitesimal increment of time.

#### Explanation of the Equation:
- **$-\frac{\partial V}{\partial t} dt$**: This term represents the loss in option value purely due to the passage of time, often referred to as time decay in options trading. As time moves closer to the option's expiration, the opportunity for the option to be profitable decreases, generally leading to a decrease in its value.
- **$-\frac{1}{2} \sigma^2 S^2 \frac{\partial^2 V}{\partial S^2} dt$**: This term accounts for the change in the portfolio’s value due to changes in volatility and the underlying asset's price, factoring in the curvature (gamma) of the option’s price response. It captures how sensitive the option's price is to large movements in the stock price.

#### Practical Significance:
- **Portfolio Management**: Understanding these components helps in managing the risk and exposure of an options-based portfolio. By hedging an option against its gamma and theta, traders can maintain a portfolio that is less sensitive to small changes in market conditions.
- **Risk Assessment**: The equation emphasizes the impact of time decay and volatility on the hedged portfolio's value. Traders use this information to adjust their strategies, particularly as options near their expiration.
- **Financial Strategy**: This simplified formula is crucial in delta-gamma hedging, where the objective is to make a portfolio immune not just to changes in the underlying asset's price but also to changes in its delta, thereby stabilizing the portfolio against shifts in market conditions.

### Return on a Risk Free Portfolio

$$
d \Pi=r \Pi d t
$$

#### Components of the Equation:
- **$d\Pi$**: Represents the infinitesimal change in the value of the portfolio.
- **$r$**: The risk-free rate of return, which is a constant. This rate is typically represented by the yield on government securities like U.S. Treasury bills, which are considered free from credit risk.
- **$\Pi$**: The current value of the portfolio.
- **$dt$**: An infinitesimal increment of time.

#### Explanation of the Equation:
- **$r \Pi dt$**: This term multiplies the risk-free rate $r$ by the current portfolio value $\Pi$ and the time increment $dt$. It calculates the amount by which a risk-free investment increases over the infinitesimal time period $dt$.

#### Practical Significance:
- **Compounding Interest**: This formula is foundational in understanding how investments that earn a constant rate of interest grow over time. The product $r \Pi$ represents the interest earned over the time period $dt$, and as time progresses, $\Pi$ increases, leading to compounding growth.
- **Valuation Models**: In finance, this formula is critical in models like the Black-Scholes for options pricing or in the valuation of fixed income securities, where the risk-free rate provides a baseline for discounting future cash flows.
- **Portfolio Strategy**: The equation provides a benchmark for comparing the performance of risky assets or investment portfolios against a risk-free asset. Investors might use this information to determine the excess return offered by riskier investments over the risk-free rate.

### Black-Scholes model for pricing European options

$$
\frac{\partial V}{\partial t}+\frac{1}{2} \sigma^2 S^2 \frac{\partial^2 V}{\partial S^2}+r S \frac{\partial V}{\partial S}-r V=0
$$

#### Explanation of the Equation:

- **Time Decay**: $\frac{\partial V}{\partial t}$ reflects how the value of the option erodes as the expiration date approaches.
- **Impact of Volatility**: The term $\frac{1}{2} \sigma^2 S^2 \frac{\partial^2 V}{\partial S^2}$ accounts for the effect of the underlying asset's volatility on the option's price, particularly capturing the risks associated with large movements in the asset price.
- **Adjustment for the Risk-Free Rate**: The $r S \frac{\partial V}{\partial S}$ term adjusts the price change for the expected return of the underlying asset if it were invested at the risk-free rate, while $r V$ discounts the current value of the option at the risk-free rate. 

#### Practical Significance:
This PDE forms the basis for the Black-Scholes formula, a theoretical estimate of the price of European-style options which allows financial analysts to estimate the fair value of options, providing a benchmark for traders.
